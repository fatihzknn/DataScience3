```python
t = np.arange(0.0, 5.0, 0.1)

plt.plot(t, (np.exp(-t)* np.cos(2- np.pi *t)), "k", t, (np.exp(-t)* np.cos(2- np.pi *t)),"bo")
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)
"""
bir kaç parametreli fonksiyonlar bu şekilde tanımlanır
def RandomFonksiyonHesapla(par1, par2):
    return par1, par2, par3
cikti1, cikti2, cikti3 = RandomFonksiyonHesapla(girdi1, girdi2)
"""
t= np.arange(0.0, 5.0, 0.1)
plt.plot(t, f(t), "k", t, f(t), "bo")
plt.axis([0,10,-1,1])
plt.show()
```

```python
mu1 = 100
sigma1 = 15
x = mu1 + (sigma1 * np.random.randn(10000))

mu2 = 110
sigma2 = 10
y = mu2 + (sigma2 * np.random.randn(10000))

# söz konusu verilerin histogramı

n, bins, patches = plt.hist(x, 100, density = 1, facecolor = "b", alpha = 0.25)
n, bins, patches = plt.hist(y, 100, density = 1, facecolor = "g", alpha = 0.25)

plt.xlabel("Veriler")
plt.ylabel("Olasılıklar")
plt.title(r"Verilerin Histogramı: $\mu_1=100,\ \mu_2=110$")
# grafik içerisindeki ok ile bir şeyi belirtmek için annotate kullanılır
plt.annotate("Gauss eğrisinin tepesi", xy=(110,0.04), xytext=(130,0.045),
            arrowprops=dict(facecolor="black",shrink=0.1))
plt.text(75, 0.025, r"$\mu_1=100,\ \sigma_1=15$")# buradaki kullanımı araştır
plt.text(120, 0.035, r"$\mu_2=110,\ \sigma_2=10$")
plt.axis([40,160,0,0.05])
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1,10,0.01)
y = np.exp(x)

plt.figure()

plt.subplot(121)
plt.plot(x,y)
plt.yscale("Linear")
plt.title("Lineer (Doğrusal)")
plt.grid(True)

plt.subplot(122)
plt.plot(x,y)
plt.yscale("Log")
plt.title("Logaritmik")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
veri = pd.read_csv("olimpiyatlar_duzenlenmis.csv")
veri.head()
```

```python
def sacilimGrafikCiz():
    sns.scatterplot(x = "boy", y = "kilo", data = veri)
    plt.title("Boy ve Ağırlık İlişkisi - Beyaz Izgara Tema")
    plt.show()
sacilimGrafikCiz()
```

```python
sns.set_style("white")
sacilimGrafikCiz()
```

```python
sns.set_style("darkgrid")
sacilimGrafikCiz()
```

```python
sns.set_style("whitegrid")
sacilimGrafikCiz()
```
