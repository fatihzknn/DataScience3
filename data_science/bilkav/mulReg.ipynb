```python
from sklearn import preprocessing
veriler = pd.read_csv("odev_tenis.csv.opdownload")
veriler2 = veriler.apply(preprocessing.LabelEncoder().fit_transform)
c = veriler2.iloc[:,:1]
ohe = preprocessing.OneHotEncoder()
c = ohe.fit_transform(c).toarray()
hava_durumu = pd.DataFrame(data = c, index = range(14), columns = ["o","r","s"])
son_veriler = pd.concat([hava_durumu,veriler.iloc[:,1:3]],axis = 1)
son_veriler = pd.concat([son_veriler,veriler2.iloc[:-2:]],axis = 1)
son_veriler.head()
```

```python
son_veriler = pd.concat([son_veriler3,son_veriler2], axis = 1)


son_veriler["windy"][13] = 1
son_veriler["play"][13] = 0
son_veriler["windy"][12] = 0
son_veriler["play"][12] = 1
son_veriler.info()
```
